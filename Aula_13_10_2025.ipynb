```python
# Faça uma lista de alunos usando append, insert, remove e pop em Python.

# Lista Inicial de Alunos
alunos = ["Thayná", "Bruna", "Lucas", "Pedro", "Rafaela", "Patrícia", "Ana", "Maria"]

# Colocando Append adicionando no fim da lista
alunos.append("Luana")
print("No fim da lista ('Luana'):", alunos)
alunos.append("Lucarion")
print("No fim da lista ('Lucarion'):", alunos)

# Insert
alunos.insert(1, "Mauro")
print("Após inserir(1, 'Mauro'):", alunos)

alunos.insert(0, "Michel")
print("Após inserir(0, 'Michel'):", alunos)

# Remove
alunos.remove("Pedro")
print("Após remover('Pedro'):", alunos)

# Pop (Removendo último elemento)
aluno_removido = alunos.pop(2)
print(f"Após remover ultimo aluno(2): {alunos} - elemento removido: '{aluno_removido}'")

ultimo_aluno = alunos.pop()
print(f"Após remover ultimo aluno(): {alunos} - Último aluno removido: '{ultimo_aluno}'")
```

```python
# Crie um algoritmo que consiga receber uma lista e utilizar 3 métodos nesta mesma lista.
# Crie a lista do que desejar
# Decida os métodos
# Crie as variáveis dos inputs e dos métodos
# Crie a saída da mensagem com as informações

# Lista de classes
classes = ["Guerreiro", "Mago", "Clérigo", "Ladino", "Bárbaro", "Bardo", "Paladino", "Ranger", "Feiticeiro", "Bruxo"]

# Função para adicionar homebrew (nome fixo: Mago Intergaláctico)
def add_homebrew():
    nova_classe = "Mago Intergaláctico"
    if nova_classe not in classes:
        classes.append(nova_classe)
        print(f"✅ {nova_classe} adicionada!")
    else:
        print("⚠️ Classe já está na lista.")

# Função para remover homebrew (nome fixo: Mago Intergaláctico)
def remove_homebrew():
    classe = "Mago Intergaláctico"
    if classe in classes:
        classes.remove(classe)
        print(f"✅ {classe} removida!")
    else:
        print("❌ Classe não encontrada.")

# Função que mostra as condições gerais
def condicoes_multiclasse():
    print("\nCONDIÇÕES:")
    print("- Ter nível 2+")
    print("- Atributos em 13+")
    print("- Cada classe tem seus requisitos")

# Função para ver uma classe específica
def verificar_classe(classe):
    requisitos = {
        "Guerreiro": "Força 13",
        "Mago": "Inteligência 13",
        "Clérigo": "Sabedoria 13",
        "Ladino": "Destreza 13",
        "Bárbaro": "Força 13",
        "Bardo": "Carisma 13",
        "Paladino": "Força 13 e Carisma 13",
        "Ranger": "Destreza 13 e Sabedoria 13"
    }

    if classe in requisitos:
        print(f"\n{classe}: {requisitos[classe]}")
    else:
        print(f"\n{classe}: Requisitos desconhecidos ou classe homebrew.")

# Função principal - ver o que você pode fazer
def ver_multiclasses():
    print("\nDigite seus atributos:")
    forca = int(input("Força: "))
    destreza = int(input("Destreza: "))
    inteligencia = int(input("Inteligência: "))
    sabedoria = int(input("Sabedoria: "))
    carisma = int(input("Carisma: "))

    print("\nVocê pode ser:")

    if forca >= 13:
        print("✅ Guerreiro")
        print("✅ Bárbaro")
    if forca >= 13 and carisma >= 13:
        print("✅ Paladino")
    if destreza >= 13:
        print("✅ Ladino")
    if destreza >= 13 and sabedoria >= 13:
        print("✅ Ranger")
    if inteligencia >= 13:
        print("✅ Mago")
    if sabedoria >= 13:
        print("✅ Clérigo")
    if carisma >= 13:
        print("✅ Bardo")

# ==========================
# Execução do sistema
# ==========================

print("=== SISTEMA DE MULTICLASSE D&D ===")

# Adiciona classe homebrew
add_homebrew()

# Exibe lista de classes atualizada
print(f"Classes: {', '.join(classes)}")

# Mostra condições gerais
condicoes_multiclasse()

# Verifica duas classes como exemplo
verificar_classe("Paladino")
verificar_classe("Mago Intergaláctico")

# Calcula possibilidades com base nos atributos
ver_multiclasses()

# Se quiser testar a remoção:
# remove_homebrew()
```

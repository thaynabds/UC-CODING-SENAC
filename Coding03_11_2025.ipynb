```python
#Aprendendo Tuplas 03.11.2025
#Métodos e funções de Tuplas (use 3)
#Tupla   turmas=('Matemática','Geografi','Biologia','Hitória','Lógica')
#Acesso por índice   turmas[0]->'Matemática'    Funciona igual ás listas
#Tentativa de alteração  turmas[1]='Geografia'  XTyppeError: tuplas são imutáveis
dias=('domingo','segunda','terça','quarta','quinta','sexta','sábado')
print(dias[5])

#Imprimindo Elementos de uma Tupla for loop + print
turmas=('Matemática','Geografi','Biologia','Hitória','Lógica') # Added the definition of 'turmas'
for e in turmas:
    print(e)

#Verificando o Tamanho de uma Tupla len()
estudante=('Eduardo','Lucas','Pedro','Miguel','Lucas','Kayo','Rayssa','Ivonalda','Eva')
print(len(estudante))

#Checando por Valores em uma Tupla
estudante=('Eduardo','Lucas','Pedro','Miguel','Lucas','Kayo','Rayssa','Ivonalda','Eva')
print('Gabriel' in estudante)
print(1990 in estudante)
print('Japão' not in estudante)
print('Brasil' not in estudante)

#Deletando a Tupla
loucos=('Eduardo','Lucas','Pedro','Miguel','Lucas','Kayo','Rayssa','Ivonalda','Eva')
del loucos

#Método count()
animais=('Lobo','Tigre','Onça','Golfinho','Gaivota','Lobo','Tigre','Tigre','Tigre')
print(animais.count('Gaivota'))
print(animais.count('Tigre'))
print(animais.count('Lobo'))

#Método index()
animais=('Lobo','Tigre','Onça','Golfinho','Gaivota','Lobo','Tigre','Tigre','Tigre')
print(animais.index('Lobo'))
print(animais.index('Onça'))

#Atualizando Tuplas
linguagens=('Python','Java','C')
print(linguagens)
linguagens[0] = "C++"
x = (1,2)
y = (3,4)
z = x + y
print(z)
```
